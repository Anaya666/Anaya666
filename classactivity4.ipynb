```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
tips = sns.load_dataset("tips")
print(tips.head(20))
```

```python
sns.boxplot(data =tips, y= "tip",x="sex",palette={"Male": "purple", "Female": "lightblue"})
plt.title("Resturaunt tips by sex")
plt.ylabel("Tip amount ")
plt.xlabel("Sex")
plt.show()
```

The graph is a boxplot which reveals the sex (male or female) on the x axis and the tip amount on the y axis. It indicates that the median tip (indicated by central line of the plot) given by males is slightly higher than the median tip given by females. The maximum tip amount given by makes is higher than the maximum tip amount given by females, but the minimum tip amount given by both genders is roughly the same. This explains why the range of tip amounts given by men is greater than the range of tip amounts given by females.

```python

sns.scatterplot(data=tips, x="tip", y="total_bill", alpha=0.6, size="size", hue="sex",palette="deep", sizes=(20, 1000))
plt.title("Total bill vs tip")
plt.xlabel("Tip amount")
plt.ylabel("Total bill")
plt.show()
```

This is a bubble plot which shows the correlation between tip amount (on the x axis) and the total bill (on the y axis). There seems to be a positive correlation between the tip amount and total bill. The hue = sex added a third variable which categorises the data by sex where blue indicates male and red/ orange indicates female. The size of the bubbles is representative of the size variable in the dataset - that is, the number of people at the table. Since size doesnt have any explicit visual correlation with tip amount and tital bill, size has no influence on either.

```python
sns.pairplot(tips, hue="time")
```

This graphic is segregated by color where blue represents lunch and orange represents dinner. For the first graph (bottom left) each size (number of people at the table) spans a range of amounts of the total_bill. This range keeps increasing. For instance, size 1 has a total bill between ~1 and ~10 usd, size 2 has a total bill between ~2 and 40 usd, size 3 has a total bill between ~10 and 45 usd and so on. This points to a positive correlation between number of people seated at the table (size) and the increasing minimum values of the range of total bills for each size category. There is also a weak positive correlation between total bill and tip amount and the tips are most highly concentrated between a total bill amount of 5 to 25 usd. Additionally, with respect to other variables there is a time based difference show different distributions, with dinner bills and tips generally being higher than lunch, suggesting that customers spend more during dinner.This is supported by the density plots which indicate that dinner has a wider spread in total bill and tip values, whereas lunch transactions tend to be lower in both aspects.A few data points suggest outliers in total bill and tip amounts, particularly at higher values, which could indicate unusually large bills or generous tipping behaviour. Further, larger groups (size column) do not necessarily result in higher tips, as the data points for different group sizes are scattered without a strong trend.

```python
#the question I want to explore is:
#How does the size of a dining party influence the total bill, and does this relationship vary between lunch and dinner?"
sns.boxplot(data=tips, x="size", y="total_bill", hue="time",palette="deep")
plt.title("size vs total bill")
plt.xlabel("size of the dining party")
plt.ylabel("Total bill")
plt.show()
```

I chose a box plot to explore this research question because it effectively displays the relationship between party size and total bill while also highlighting group differences between lunch and dinner. The graph suggests a weak positive correlation between party size and total bill, as indicated by the generally increasing median values across party sizes, though with some inconsistencies.
Notably, for party size = 1, the median total bill is higher for lunch than dinner, whereas for party size = 2, the median total bill is higher for dinner. At party size = 6, the trend reverses again, with lunch having a higher median total bill. This fluctuating pattern suggests no clear correlation between meal time (lunch or dinner) and total bill with respect to party size. The overlapping interquartile ranges and outliers further support the lack of a strong trend, indicating that other factors beyond party size and meal time likely influence total spending.
