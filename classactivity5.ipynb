```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px #allows you to create interactive plots
```

```python
penguin_data = sns.load_dataset('penguins')
print(penguin_data.head())
```

```python
sns.kdeplot(data = penguin_data, x="body_mass_g", hue= "species", fill=True)
```

This graph reveals a density plot shows the distribution of body mass for different penguin species. The different peaks of the three different graphs )(representative of their respective species) indicate the most common body mass range for that species. Adelie penguines and Chinstrap penguines have similar ranges of bodymasses while Gentoo penguines have a higher range.
This density plot helps us understand how body mass varies among the three species, potentially reflecting ecological and biological differences. It also provides an intuitive way to compare species without relying on summary statistics alone. By analyzing the density distribution, researchers can investigate factors affecting penguin body mass, such as diet, climate conditions, and migratory behavior.

```python
sns.scatterplot(data= penguin_data, x="bill_length_mm", y="flipper_length_mm", size= "body_mass_g", hue = "species", sizes=(20,700), alpha = 0.5)
plt.legend(bbox_to_anchor= (1.5, 1))
plt.xlabel("Length of Beak")
plt.ylabel("Length of flippers")
plt.title("Penguin measurements across three species")
```

While it successfully presents multiple variables—beak length, flipper length, body mass, and species—it may overwhelm a general audience unfamiliar with complex data visualizations. The inclusion of bubble size as an additional dimension adds to the cognitive load, making it difficult for readers to quickly grasp the key insights. Additionally, overlapping bubbles obscure important data points, reducing the overall readability. The use of multiple colors helps differentiate species, but the sheer number of data points creates visual clutter, making the distinctions between species less effective.
