```python
# Vectors v and w
v = np.array([1, 2, 3, 4])
w = np.array([0, 1, 0, 1])
```

```python
# let's calculate the l1 norm of v and w
### FILL IN THE ORDER FOR EACH OF THE STATEMENTS BELOW ###
print('V L1 Norm:', LA.norm(v, ord = 1))
print('W L1 Norm:', LA.norm(w, ord = 1))
print('V L2 Norm:', LA.norm(v, ord = 2))
print('W L2 Norm:', LA.norm(w, ord = 2))


#ord =1 also known as the manhattan norm- sum of the absolute values of each element
#ord =2 is also known as the euclidean norm-It is the square root of the sum of the squared elements of the vector:
#ord helps you determine the magnitude or length of the vector accordingly
```

```python
import matplotlib.pyplot as plt
# n is the range of p values
n = 100

# show the u vector
print('The U Vector:', u)
#u is the vector whose p norms will be calculated

# create the arrays for the the p values to use
p_vals = np.arange(-n, n, step = 1)
#p_vals: Creates an array of integers from -100 to 100 (excluding 100) which will be the range
#of p- values to test

# create an empty array to fill the resulting norms with
p_norms = np.empty(len(p_vals)) #Creates an empty array to store the computed norms.

# calcualte the p norm for each value of p
for i, val in enumerate(p_vals):
  p_norms[i] = LA.norm(u, ord = val)

#This calculates the p-norm of the vector u for each p-value in p_vals and stores the result in p_norms.

# plot the value using matplotlib
plt.plot(
    p_vals,
    p_norms,
    color = 'black'
)

# see if it converges to the max value as it goes to infinity
absolute_max = np.abs(np.max(u))

# add labels and show
plt.xlabel('P Value')
plt.ylabel('P Norm Value')
plt.title(r'What happens to $\ell_p$ as p $\rightarrow$ $\infty$?')
plt.axhline(y = absolute_max, linestyle = '--', color = 'red')
plt.show()

#This code explores the behavior of the p norm (generalized p norm) as the value of p changes, ranging from negative to
#positive infinity



```

```python
# Transpose of Matrix
V = np.array(
    [
        [1, 2, 3],
        [4, 5, 6]
    ]
)

# all numpy arrays have a shape attribute that we can access
print('Shape of V:', V.shape, '\n')

### FIND THE TRANSPOSE OF V AND COMPARE IT'S SHAPE TO V (using numpy transpose function)
# Find the transpose of V
V_transpose = np.transpose(V)

# Print the transpose and its shape
print('Transpose of V:\n', V_transpose, '\n')
print('Shape of Transpose of V:', V_transpose.shape)
```

```python
# Transpose of a Matrix without Numpy

def transpose(matrix):
    """Transposes a given matrix."""

    # Create a new matrix with swapped dimensions
    transposed_matrix = [[0 for _ in range(len(matrix))] for _ in range(len(matrix[0]))]

    # Fill the transposed matrix
    for i in range(len(matrix)):
        for j in range(len(matrix[0])):

            # swapping the j,i entry of the transpose to the i,j entry of the original matrix
            transposed_matrix[j][i] = matrix[i][j]

    return transposed_matrix

matrix = V
print(transpose(matrix))
```

```python
### we can multiply a vector and a matrix together

# let's make an array to multiply with our V matrix
# it must be 1x2 or 3x1 to be multiplied with a 2x3 matrix
v1 = np.array([1, 2])

# the transpose makes it a column vector
v2 = np.array([1, 2, 3]).T


### CALCULATE THE MATRIX MULTIPLICATION OF v1 dot V and V dot v2
v1_dot_V = v1 @ V  # Row vector multiplied by a 2x3 matrix
V_dot_v2 = V @ v2  # 2x3 matrix multiplied by a column vector

# Display results
print("Result of v1 dot V:\n", v1_dot_V)
print("\nResult of V dot v2:\n", V_dot_v2)

```

```python
# how about matrix multiplication?
# define a valid matrix
U = np.array(
    [
        [5, 6],
        [2, 7],
        [10, 3]
    ]
)

print('U Matrix', '\n', U, '\n')

### CALCULATE THE MATRIX MULTIPLICATION OF UV AND VU (using matmul)
# Calculate matrix multiplication
UV = np.matmul(U, V)  # U (3x2) dot V (2x3) results in a 3x3 matrix
VU = np.matmul(V, U)  # V (2x3) dot U (3x2) results in a 2x2 matrix

# Display results
print("Matrix Multiplication UV:\n", UV, "\n")
print("Matrix Multiplication VU:\n", VU, "\n")
```
